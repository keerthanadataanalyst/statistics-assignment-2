```python
#16. Create a confidence interval for a dataset using Python and interpret the result.
```

A. A confidence interval gives a range of values that is likely to contain the population parameter (e.g., population mean) with a certain level of confidence. For example, a 95% confidence interval suggests that we are 95% confident the population mean falls within this range.

```python
import numpy as np
import scipy.stats as stats

#sample data - scores
data = np.random.normal(loc=50, scale=10, size=5)

#calculate mean, std dev
sample_mean = np.mean(data)
sample_std_dev = np.std(data, ddof =1 )  # ddof=1 for sample standard deviation
n = len(data)

#confidence level(95%)

confidence_level = 0.95
alpha = 1 - confidence_level

#z scores for 95% confidence
z_critical = stats.norm.ppf(1 - alpha / 2)

#standard error

se = sample_std_dev/ np.sqrt(n)

# margin of error 

margin_of_error = z_critical*se

#confidence_interval

lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

#interpret results

print("sample mean is", sample_mean )
print("sample standard deviation is", sample_std_dev )
print("z-critical is", z_critical )
print("margin of error is", margin_of_error)
print("confidence_interval is", (lower_bound, upper_bound))

# Interpretation
print(f"\nInterpretation: We are 95% confident that the true population mean lies between {lower_bound:.2f} and {upper_bound:.2f}.")
```

```python
#22. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution5
```

```python
'''
Range - 

The range is the difference between the maximum and minimum values in a dataset:

Range= Max(X) − Min(X)

Interquartile range - 

The IQR is the range between the first quartile (Q1) and the third quartile (Q3).
It is the spread of the middle 50% of your data.

IQR=Q3−Q1

Where:
Q1 is the 25th percentile (25% of the data is below this value).
Q3 is the 75th percentile (75% of the data is below this value).

'''
```

```python
#first part of question - Generate dataset from nomal distribution

np.random.seed(0)
data = np.random.normal(50,10,1000) # Mean=50, Std Dev=10, 1000 samples

#calculate range

range_ = np.max(data) - np.min(data)

#calculate IQR
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

print("Range of the dataset:", range_)
print( "IQR of the dataset:", IQR)

```

```python
#23. Implement Z-score normalization on a dataset and visualize its transformation
```

```python
'''
Z-score normalization (or standardization) is a method used to transform the data such that
the mean becomes 0 and the standard deviation becomes 1. 

The formula for Z-score normalization is: Z = data point - mean/ sigma

'''
```

```python
#generate a dataset
np.random.seed(0)
data = np.random.normal(50,10,1000)  # Mean=50, Std Dev=10, 1000 samples

#calculate mean and std dev
mean = np.mean(data)
std_dev = np.std(data)

#z score normalization
normalized_data = (data - mean /std_dev)

#plotting graphs

plt.figure(figsize=(12, 6))

#original data
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.title("Original Data Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)

#normalized data
plt.subplot(1, 2, 2)
plt.hist(normalized_data, bins=30, color='lightgreen', edgecolor='black', alpha=0.7)
plt.title("Normalized Data Distribution (Z-Score)")
plt.xlabel("Z-Score Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
